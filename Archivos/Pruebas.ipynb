```python
import pandas as pd
```

## Cargar todos los matrimonios

```python
mat09 = pd.read_spss('Matrimonios_2009.sav')
mat10 = pd.read_spss('Matrimonios_2010.sav')
mat11 = pd.read_spss('Matrimonios_2011.sav')
mat12 = pd.read_spss('Matrimonios_2013.sav')
mat13 = pd.read_spss('Matrimonios_2013.sav')
mat14 = pd.read_spss('Matrimonios_2014.sav')
mat15 = pd.read_spss('matrimonios_2015.sav')
mat16 = pd.read_spss('matrimonios_2016.sav')
mat17 = pd.read_spss('matrimonios_2017.sav')
mat18 = pd.read_spss('matrimonios_2018.sav')
mat19 = pd.read_spss('matrimonios_2019.sav')
mat20 = pd.read_spss('matrimonios_2020.sav')
```

## Cargar todos los divorcios

```python
div09 = pd.read_spss('Divorcios_2009.sav')
div10 = pd.read_spss('Divorcios_2010.sav')
div11 = pd.read_spss('Divorcios_2011.sav')
div12 = pd.read_spss('Divorcios_2013.sav')
div13 = pd.read_spss('Divorcios_2013.sav')
div14 = pd.read_spss('Divorcios_2014.sav')
div15 = pd.read_spss('divorcios_2015.sav')
div16 = pd.read_spss('divorcios_2016.sav')
div17 = pd.read_spss('divorcios_2017.sav')
div18 = pd.read_spss('divorcios_2018.sav')
div19 = pd.read_spss('divorcios_2019.sav')
div20 = pd.read_spss('divorcios_2020.sav')
```

## unir los dataframes

```python
framesDiv = [div09, div10, div11, div12,div13,div14,div15,div16,div17,div18,div19,div20]
framesMat = [mat09, mat10, mat11, mat12,mat13,mat14,mat15,mat16,mat17,mat18,mat19,mat20]
```

```python
#lower the name of columns for all dataframes inside this list
for df in framesDiv:
    df.columns = df.columns.str.lower()
```

```python
#lower the name of columns for all dataframes inside this list
for df in framesMat:
    df.columns = df.columns.str.lower()
```

```python
df_matrimonios = pd.concat(framesMat)
df_divorcios = pd.concat(framesDiv)
```

```python
df_matrimonios.to_csv('matrimonios.csv',index=False,encoding='utf-8')
df_divorcios.to_csv('divorcios.csv',index=False,encoding='utf-8')
```
